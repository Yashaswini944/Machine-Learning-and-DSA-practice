```python
import pandas as pd
```

```python
# construct Dataframe from a dictionary of lists

data = {'ProductName':['Product A', 'Product B','Product C'], 'ProductPrice':[22250, 16609, 13590]}
df = pd.DataFrame(data) #passing arguments
df
```

```python
#construct Dataframe from a dictionary of lists + specify index
data = {'ProductName':['Product A', 'Product B','Product C'], 'ProductPrice':[22250, 16609, 13590]}
product_IDs = ['A','B', 'C']
df = pd.DataFrame(data, index = product_IDs) #passing arguments
df
```

```python
#construct Dataframe from a list of dictionary
data = [{'ProductName':'Product A', 'ProductPrice':16609},
        {'ProductName':'Product B', 'ProductPrice': 22500},
        {'ProductName':'Product C', 'ProductPrice': 13590}]
df = pd.DataFrame(data) #passing arguments
df
```

```python
data = [{'ProductName':'Product A', 'ProductPrice':16609},
        {'ProductName':'Product B', 'ProductPrice': 22500},
        {'ProductName':'Product C', 'ProductPrice': [45015, 13590]}]
df = pd.DataFrame(data) #passing arguments
df
```

```python
data = [{'ProductName':'Product A', 'ProductPrice':16609},
        {'ProductName':'Product B', 'ProductPrice': 22500},
        {'ProductName':'Product C', 'ProductPrice': 13590},
        {'ProductName' : 'Product D'}]
df = pd.DataFrame(data) #passing arguments
df
```

```python
# construct a dataframe from a dictionary of pandas of pandaas Series

s_products = pd.Series(['Product A', 'Product B','Product c'])
s_prices = pd.Series([30098, 16677, 14500])
```

```python


data = {'ProductName':s_products, 'ProductPrice':s_prices}
df = pd.DataFrame(data)
df
```

```python
s_products = pd.Series(['Product A', 'Product B','Product c'], index = ['A', 'B','C'])
s_prices = pd.Series([30098, 16677, 14500], index = ['A', 'B','C'])

data = {'ProductName':s_products, 'ProductPrice':s_prices}
df = pd.DataFrame(data)
df
```

```python
s_products = pd.Series(['Product A', 'Product B','Product c'], index = ['A', 'B','C'])
s_prices = pd.Series([30098, 16677, 14500], index = ['C', 'A', 'B',])

data = {'ProductName':s_products, 'ProductPrice':s_prices}
df = pd.DataFrame(data)
df
```
