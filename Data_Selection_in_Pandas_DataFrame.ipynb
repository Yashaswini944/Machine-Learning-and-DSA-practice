# Data Selection

data selection or subset selection in a pandas DataFrame means extracting elements,rows,columns or subsets from such an object

-it allows us to work on just a portion of a dataset

**Indexing**

using 1 or both type of indexes a DataFrame has - the Row index & the Cloumn index- to access or select specific parts of the data
